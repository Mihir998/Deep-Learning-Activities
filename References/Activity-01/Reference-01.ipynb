```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
iris_data = load_iris()
```

```python
x = iris_data.data
y_ = iris_data.target.reshape(-1,1) # _ as there is no one hot encoder
```

```python
y_
```

```python
encoder = OneHotEncoder(sparse=False)
y = encoder.fit_transform(y_)
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.20)
```

```python
model = tf.keras.Sequential([
                             tf.keras.layers.InputLayer(input_shape=(4, )),
                             tf.keras.layers.Dense(10, activation='relu'),
                             tf.keras.layers.Dropout(0.05), #removes all the values below 0.05
                             tf.keras.layers.Dense(3, activation='softmax') #since we have 3 output variable as seen above
])

model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.006), loss='categorical_crossentropy', metrics=['accuracy']) #categorical_crossentropy is used because we have categorical data
model.summary()
```

```python
history = model.fit(x_train, y_train,epochs=100)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(range(100), history.history['loss'])
plt.plot(range(100), history.history['accuracy'])
plt.legend()
plt.show()
```

```python
model.evaluate(x_test, y_test)
```
